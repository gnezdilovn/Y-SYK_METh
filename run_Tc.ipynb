```python
def ebc_array(logt_max_plus):
    f = []
    for r in range(logt_max_plus):
        if r == 0:
            g = [0.05, 0.06, 0.07, 0.08, 0.09] 
        else:    
            g = [0.01 * i * 10 ** r for i in range(1, 10)]
        f = f + g
    f = f + [10 ** (logt_max_plus - 2)]    
    return np.round(f, 3) 
```

```python
# modified T_c solver (includes the convergence cutoff on number of Matsubara frequencies)
def T_c_conv(a, g_array):
    f = np.zeros(len(g_array))
    for k in range(len(g_array)):
        g_k = g_array[k] 
        #if g_k <= 1.:
        #    if a >= 1.:
        #        N2 = 1500
        #    else:    
        #        N2 = 300
        if g_k < 1.:
            if a >= 1.:
                N2 = 300
            else:    
                N2 = 100
        else:
            N2 = 100    
        Ts_k = T_array(g_k, a)
        Det_k = np.zeros(len(Ts_k))
        print('g={}'.format(g_k))
        print('N2={}'.format(N2))
        for t in range(len(Ts_k)):
            T = Ts_k[t]
            print('T={}'.format(T))
            Det_k[t] = Det(N2, T, g_k, a)
            print(Det_k[t])
            if t > 0:
                if  Det_k[t] * Det_k[t-1] < 0:
                    if np.abs(Det_k[t-1]) < np.abs(Det_k[t]): 
                        print('T_c={}'.format(Ts_k[t-1]))
                        f[k] = Ts_k[t-1]
                    else: 
                        print('T_c={}'.format(Ts_k[t]))
                        f[k] = Ts_k[t]
                    break                    
    return f 
```
